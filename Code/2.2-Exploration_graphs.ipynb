```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('daily_aggregate_data.csv')
data['Date'] = pd.to_datetime(data.Date)
data.head()
```

```python
corr_mat = data.corr().values
data.corr()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(15,7))
props = dict(boxstyle='round', facecolor='wheat', alpha=0.9)

ax1.scatter(data.mean_cloud, data.mean_power, c='darkblue')
ax1.set_xlabel('Mean Cloud Cover', fontsize=14, fontweight='bold')
ax1.set_ylabel('Mean Power (KW)', fontsize=14, fontweight='bold')
ax1.text(0.05, 0.95, f'Corr: {round(corr_mat[4,0],2)}', transform=ax1.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

ax2.scatter(data.mean_temp, data.mean_power, c='darkblue')
ax2.set_xlabel('Mean Temperature (F)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Mean Power (KW)', fontsize=14, fontweight='bold')
ax2.text(0.05, 0.95, f'Corr: {round(corr_mat[2,0],2)}', transform=ax2.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

ax3.scatter(data.mean_radiation, data.mean_power, c='darkblue')
ax3.set_xlabel('Mean Solar Radiation (KW/m^2)', fontsize=14, fontweight='bold')
ax3.set_ylabel('Mean Power (KW)', fontsize=14, fontweight='bold')
ax3.text(0.05, 0.95, f'Corr: {round(corr_mat[1,0],2)}', transform=ax3.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

plt.show()
```
